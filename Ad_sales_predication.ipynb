```python
data = pd.read_csv('purchase_records.csv')
print(data.shape)
data.head(5)
```

```python
x= data.iloc[:,:-1]
x
```

```python
y= data.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.25,random_state=0)
print(y_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train = sc.fit_transform(x_train)

print(x_train)
```

```python
x_test = sc.transform(x_test)
```

```python
#print(x_train)
from sklearn.linear_model import LogisticRegression
#x_train = np.reshape(300,2)
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
```

```python
age = int(input('Enter the age of customer:'))
salary = int(input("Enter the salart of customer"))
newcustomer=[[age,salary]]
result = model.predict(sc.transform(newcustomer))
print(result)
if result == 1:
  print('Customer will Buy')
else:
  print("Customer won'n Buy")
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
cm=confusion_matrix(y_test,ypred)

print('Confusion matrix:\n',cm)
acc= accuracy_score(y_test,ypred)
print("Accuracy Score {0}%".format(acc*100))
```
